### Calculating Observational Monthly Transport Averages

First, I have to make a function where I can give it a certain depth range, and it will give me the observational data transport over the time I'm interested in for that depth range (so the output should be a 1d array).
<br>
<br> In this function, I will give it the along strait velocities, starting depth and ending depth that I want, along with the xdist in order to do the horizontal trapezoidal integration.

I first consructed an array that stores the 12 months I'm interested in to create monthly averages of the transport data stored in the observational data.

```python
mon_times = []   #21 months total
for i in range(2012,2015):
    for j in range(1,13):
        if i == 2012 and j < 7:
            pass
        elif i == 2014 and j > 3:
            pass
        else:
            mon_times = np.append(mon_times, dt.datetime(i,j,1))
```

Next, I am storing the indices of the times from the observational data that match with the start date as well as the first of each month so that I can then average all the trasnport values for each day in each month I'm interested in.

```python
indices = []
for i in range(len(obstime)):
    if obstime[i] == dt.datetime(2012,7,23):
        indices = np.append(indices, i)
    for j in range(2012,2015):
        for k in range(1,13):
            if obstime[i] == dt.datetime(j,k,1):
                indices = np.append(indices, i)
```

Now, I'm making a function to calculate the monthly transport values for specific depth ranges I'm interested in. I specify the depths I'm interested in, average the ASV values across the months, integrate trapezoidally across the points along the cross section, multiply by 20 (because each grid cell is 20 m deep), sum along the depth dimension, and lastly change the units to Sverdrups.

```python
def calc_obs_trans(asv,start_dp,end_dp,depth,xdist,indices):
    # first figure out chunk of depths I'm interested in
    dp = np.where((depth >= start_dp) & (depth <= end_dp))
    asv_dp = asv[dp]
    
    # average the asv values over months
    asv_mon = np.zeros((dp[0].shape[0],asv.shape[1],indices.shape[0]))        # asv_mon is where i will store the monthly averaged asv's
    for i in range(len(indices)):
        ind = int(indices[i])
        if ind == 586:
            mon_mean = asv_dp[:,:,586:].mean(axis=2)
            asv_mon[:,:,i] = mon_mean
        else:
            ind2 = int(indices[i+1])
            mon_mean = asv_dp[:,:,ind:ind2].mean(axis=2)
            asv_mon[:,:,i] = mon_mean
    
    # do a trapezoidal integration along the strait for each month
    asv_trapz = np.zeros((dp[0].shape[0],indices.shape[0]))
    for i in range(dp[0].shape[0]):
        for j in range(indices.shape[0]):
            temp = asv_mon[i,:,j]
            temptrapz = np.trapz(temp, xdist)
            asv_trapz[i,j] = temptrapz
    
    # multiply the whole matrix by 20 for the depth of each cell & sum along the depth dimension
    transport = asv_trapz.sum(axis=0)*20/10**6
    
    return(transport)
```

```python
obs_transport = calc_obs_trans(asv,100,850,depth,xdist,indices)
obs_transport
```

## 5. Surface Area Calculations
I want to use the udata and vdata to the use the corresponding 'x' and 'y' points to save the side lengths into a vector. I then need to use those side lengths and multiply by the thickness of each cell (which I calculate using my 'find_dz' function) to get the surface areas for each grid cell in the cross section that I'm interested in. Once I have this information, I can calculate transport the way I wanted to before.

Make vector 'dsides' that stores the different lengths from i,j-point to i,j-point to then multiply by depths to get SA values.

```python
dsides = []
for d in data:
    dtype,i,j,lon,lat = d
    i = int(i)
    j = int(j)
    if dtype == 0:
        dy = dyCu.isel(xq=i,yh=j).values
        dsides = np.append(dsides, dy)
    else:
        dx = dxCv.isel(xh=i,yq=j).values
        dsides = np.append(dsides, dx)
```

Now, I need to calculate the depth variables for each point.

```python
deepest = []
for d in data:
    dtype,i,j,_,_ = d
    i,j = int(i),int(j)
    depth = o_s['deptho'].isel(xh=i,yh=j).values
    deepest = np.append(deepest,depth)
```

With 'deepest', I am finding the deepest measurement I can count as the ocean floor for each point in 'data'. This will help me in my function to find the right SA values.

```python
lev = grid['lev']
```

Since 'lev' contains values of the midpoints of the grid cells, I need the function 'find_dz' to tell me how deep each grid cell is.

```python
def find_dz(lev):
    mylev = []
    depth = 0
    for i in range(len(lev)):
        dz = abs(depth-lev[i])*2
        depth += dz
        mylev = np.append(mylev,dz)
    return mylev
```

The function 'find_dz' finds the dz/dr, or the depth of each grid cell.

```python
mylev = find_dz(lev[:30])
```

I want to make an array (call it 'zl') that stores the values of the edges of the grid boxes in the vertical direction.

```python
zl = np.array([0])
dpth = 0
for i in mylev:
    dpth += i
    zl = np.append(zl, dpth)
```

I want to make a (point x depth) array now that stores in all the 'dz' that correspond to each point in the cross section.

```python
SA = np.empty((dsides.shape[0], mylev.shape[0]))
```

I modified the original function to allow me to choose what depth level to stop at. The elif statement allows me to select a fraction of the last grid cell that partially has water in it so that I can get a better estimate of the total transport.

```python
#still have to modify to reflect 
def calc_SA_specificdepth(SA,dsides,mylev,lev,deepest,shallowest):
    for i in range(len(dsides)):
        for j in range(len(mylev)):
            if shallowest < lev[j] < deepest:
                SA[i][j] = dsides[i] * mylev[j]
            elif shallowest < lev[j-1] < deepest and lev[j] > deepest:
                dif = deepest - lev[j-1].values
                SA[i][j] = dsides[i] * dif
            else:
                SA[i][j] = 0
    return SA
```

```python
SA850 = calc_SA_specificdepth(SA,dsides,mylev,lev,850,100)
```

## 6. Calculating Transport from this SA data

I changed the transport calculation function once again to have a 3d matrix of velocity indexed by points along the cross section, time, and depth points. This takes about a minute to make. After I make this matrix, I can then do calculations much more easily and quickly.

```python
def make_velmat(ds,data,start,end,dp,times):
    vel_mat = np.zeros((data.shape[0],times.shape[0],dp))
    for d in range(len(data)):
        dtype,i,j,lon,lat = data[d]
        i,j=int(i),int(j)
        if dtype == 0:
            vel_mat[d] = ds['uo'].isel(member_id=0,x=i,y=j,time=slice(times[0],times[-1]+1),lev=slice(start,end+1))
        else:
            vel_mat[d] = ds['vo'].isel(member_id=0,x=i,y=j,time=slice(times[0],times[-1]+1),lev=slice(start,end+1))
    return vel_mat
```

'times' is an array that stores the time indices I'm interested in that I can use to plug into the transport calculation function while 'times1' is the array I will use to actually plot the transport values.
