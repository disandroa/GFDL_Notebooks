### Calculating Observational Monthly Transport Averages

First, I have to make a function where I can give it a certain depth range, and it will give me the observational data transport over the time I'm interested in for that depth range (so the output should be a 1d array).
<br>
<br> In this function, I will give it the along strait velocities, starting depth and ending depth that I want, along with the xdist in order to do the horizontal trapezoidal integration.

I first consructed an array that stores the 12 months I'm interested in to create monthly averages of the transport data stored in the observational data.

```python
mon_times = []   #21 months total
for i in range(2012,2015):
    for j in range(1,13):
        if i == 2012 and j < 7:
            pass
        elif i == 2014 and j > 3:
            pass
        else:
            mon_times = np.append(mon_times, dt.datetime(i,j,1))
```

Next, I am storing the indices of the times from the observational data that match with the start date as well as the first of each month so that I can then average all the trasnport values for each day in each month I'm interested in.

```python
indices = []
for i in range(len(obstime)):
    if obstime[i] == dt.datetime(2012,7,23):
        indices = np.append(indices, i)
    for j in range(2012,2015):
        for k in range(1,13):
            if obstime[i] == dt.datetime(j,k,1):
                indices = np.append(indices, i)
```

Now, I'm making a function to calculate the monthly transport values for specific depth ranges I'm interested in. I specify the depths I'm interested in, average the ASV values across the months, integrate trapezoidally across the points along the cross section, integrate trapezoidally across the depth points, and lastly change the units to Sverdrups. I made this function from a combination of the ASV_tplot.ipynb and model_v_obs.ipynb notebooks.

```python
def calc_obs_trans(asv,depth,xdist,indices):
    
    # average the asv values over months
    asv_mon = np.zeros((depth.shape[0],asv.shape[1],indices.shape[0]))        # asv_mon is where i will store the monthly averaged asv's
    for i in range(len(indices)):
        ind = int(indices[i])
        if ind == 586:
            mon_mean = asv[:,:,586:].mean(axis=2)
            asv_mon[:,:,i] = mon_mean
        else:
            ind2 = int(indices[i+1])
            mon_mean = asv[:,:,ind:ind2].mean(axis=2)
            asv_mon[:,:,i] = mon_mean                  # still m/s
    
    # do a trapezoidal integration along the strait for each month
    asv_trapz = np.zeros((depth.shape[0],indices.shape[0]))
    for i in range(depth.shape[0]):
        for j in range(indices.shape[0]):
            temp = asv_mon[i,:,j]
            temptrapz = np.trapz(temp, xdist)
            asv_trapz[i,j] = temptrapz                 # m^2/s now
    
    # integrate trapezoidally across depths
    transport = []
    for i in range(indices.shape[0]):
        ttrapz = np.trapz(asv_trapz[:,i],depth)
        transport = np.append(transport,ttrapz)
    
    # change units to Sv
    trans_sv = transport/10**6
    
    return(trans_sv)
```

```python
obs_transport = calc_obs_trans(asv,depth,xdist,indices)
obs_transport
```

## 5. Surface Area Calculations
I want to use the udata and vdata (hopefully the i and j will alternate in the way they're supposed to) to the use the corresponding 'x' and 'y' points to save the side lengths into a vector. I then need to use those side lengths and multiply by 'thkcello' (that's actually missing from this) to get the surface areas for each grid cell in the cross section that I'm interested in. Once I have this information, I can calculate transport the way I wanted to before.

Make vector 'dsides' that stores the different lengths from i,j-point to i,j-point to then multiply by depths to get SA values.

```python
dsides = []
for d in data:
    dtype,i,j,lon,lat = d
    i = int(i)
    j = int(j)
    if dtype == 0:
        dy = dyCu.isel(xq=i,yh=j).values
        dsides = np.append(dsides, dy)
    else:
        dx = dxCv.isel(xh=i,yq=j).values
        dsides = np.append(dsides, dx)
dsides
```

Now, I need to calculate the depth variables for each point.

```python
deepest = []
for d in data:
    dtype,i,j,lon,lat = d
    i,j = int(i),int(j)
    depth = o_s['deptho'].isel(xh=i,yh=j).values
    deepest = np.append(deepest,depth)
deepest
```

```python
lev = grid['lev']
lev.values
```

```python
def find_dz(lev):
    mylev = []
    depth = 0
    for i in range(len(lev)):
        dz = abs(depth-lev[i])*2
        depth += dz
        mylev = np.append(mylev,dz)
    return mylev
```

```python
mylev = find_dz(lev[:30])
mylev.shape
```

I want to make a (point x depth) array now that stores in all the 'dz' that correspond to each point in the cross section.

```python
SA = np.empty((dsides.shape[0], mylev.shape[0]))
```

```python
def calc_SA(SA, dsides, mylev, deepest, lev):
    for i in range(len(dsides)):
        deep = deepest[i]
        for j in range(len(mylev)):
            if np.isnan(deep):
                SA[i][j] = 0
            elif lev[j] < deep:
                SA[i][j] = dsides[i] * mylev[j]
            elif lev[j-1] < deep and lev[j] > deep:
                dif = deep - lev[j-1].values
                SA[i][j] = dsides[i] * dif
            else:
                SA[i][j] = 0
    return SA
```

```python
SAcalculated = calc_SA(SA,dsides,mylev,deepest,lev)
SAcalculated.shape
```

## 6. Calculating Transport from this SA data

Using matrix calculation mode again to calculate transport once again (instead of using a bunch of for loops).

```python
def make_velmat(ds,data,dp,times):
    vel_mat = np.zeros((data.shape[0],times.shape[0],dp))
    for d in range(len(data)):
        dtype,i,j,lon,lat = data[d]
        i,j=int(i),int(j)
        if dtype == 0:
            vel_mat[d] = ds['uo'].isel(member_id=0,x=i,y=j,time=slice(times[0],times[-1]+1),lev=slice(0,dp))
        else:
            vel_mat[d] = ds['vo'].isel(member_id=0,x=i,y=j,time=slice(times[0],times[-1]+1),lev=slice(0,dp))
    return vel_mat
```

```python
times = np.arange(1611,1971)
times.shape
```

```python
trans_mat3d = vel_mat3d
for i in range(vel_mat3d.shape[1]):
    trans_mat3d[:,i] = SAcalculated * vel_mat3d[:,i]
trans_mat3d[np.isnan(trans_mat3d)] = 0
```
